```python
for file in files:
    img = plt.imread(file)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

for i, file in enumerate(files[:4]):
    img = plt.imread(file)
    ax[i // 2, i % 2].imshow(img)
    ax[i // 2, i % 2].axis('off')
plt.tight_layout()
plt.savefig('./images/combined_image_1.png')
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
for i, file in enumerate(files[4:8]):
    img = plt.imread(file)
    ax[i // 2, i % 2].imshow(img)
    ax[i // 2, i % 2].axis('off')
plt.tight_layout()
plt.savefig('./images/combined_image_2.png')
plt.show()
```
