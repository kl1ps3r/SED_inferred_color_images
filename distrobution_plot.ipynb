```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
rng = np.random.default_rng()

n = np.exp(rng.normal(loc=0.2, scale=1.0, size=10000)) + 0.2
```

```python
plt.hist(n, bins=30, density=True, alpha=0.7, color='blue')
plt.xlabel('n')
plt.ylabel('Density')

plt.xscale('log')

plt.show()
```
